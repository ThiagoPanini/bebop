# Contexto: Trainer

Em geral, em um fluxo de desenvolvimento de um modelo preditivo, uma série de etapas são cumpridas de modo a treinar, avaliar e selecionar o melhor modelo para uma determinada tarefa. Considerando um pipeline já consolidado de preparação dos dados, a modelagem exige um a aplicação de uma série de funções e métodos para que novos insights possam ser retirados do processo de treinamento, fazendo com que essa se torne uma etapa extensa e trabalhosa para o cientista/desenvolvedor.

Nesse contexto, será apresentado abaixo o módulo `pycomp.ml.trainer` como uma forma dinâmica, rápida e eficiente de encapsular todos os principais métodos de treinamento e avaliação de modelos preditivos. Através de sua classe `ClassificadorBinario`, é possível construir modelos de classificação binária em poucas linhas de código, permitindo uma rica análise do melhor classificador para uma determinada tarefa.

será utilizada a base de dados [Titanic](https://www.kaggle.com/c/titanic): um problema de negócio clássico que envolve o treinamento de modelos de Machine Learning para o entendimento das principais características relacionadas a sobrevivência dos passageiros e tripulantes no naufrágio do icônico navio.

As bases estão disponíveis em dois arquivos (`train.csv` e `test.csv`) e, nas células a seguir, serão utilizadas as funções e classes presentes nos módulos `pycomp.ml.transformers` para a exploração e construção de um fluxo completo de preparação de dados para a futura utilização de modelos preditivos.

```python
# Importando bibliotecas
import os
import pandas as pd
import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Definindo variáveis do projeto
DATA_PATH = 'titanic_data/'
TRAIN_FILENAME = 'train.csv'
TEST_FILENAME = 'test.csv'
```

# Lendo Base

```python
# Lendo base de treino e verificando conteúdo
df_train = pd.read_csv(os.path.join(DATA_PATH, TRAIN_FILENAME))
print(f'Dimensões da base de treino: {df_train.shape}')
df_train.head()
```

Uma vez lida a base de treino, é possível analisar os atributos contidos na mesma a partir de seus [metadados](https://www.kaggle.com/c/titanic/data):

- **_PassengerId_:** id do passageiro ou tripulante do navio;
- **_Survived_:** variável resposta indicando a sobrevivência do passageiro ou tripulante;
- **_Pclass_:** categoria indicativa da classe do ticket do passageiro (1=alta, 2=média ou 3=baixa);
- **_Name_:** nome do passageiro ou tripulante;
- **_Sex_:** gênero do passageiro ou tripulante;
- **_Age_:** idade do passageiro ou tripulante;
- **_SibSp_:** quantidade de irmãos ou cônjuges presentes na embarcação;
- **_Parch_:** quantidade de pais ou filhos presentes na embarcação;
- **_Ticket_:** número referente ao ticket do passageiro ou tripulante;
- **_Fare_:** valor da tarifa paga pelo passageiro ou tripulante;
- **_Cabin_:** número da cabine do passageiro ou tripulante;
- **_Embarked_:** porto de embarque do passageiro ou tripulante (C=Cherbourg, Q=Queenstown, S=Southampton).

# Data Prep: Transformers

Nas documentações do pacote `pycomp`, será possível encontrar um jupyter notebook altamente detalhado com algumas das principais funcionalidades do módulo `ml.transformers`, sendo este responsável por encapsular classes prontas para a inclusão em um pipeline de preparação dos dados. 

Para facilitar a demonstração do módulo `ml.trainer`, será construído um pipeline completo de preparação para que a base original possa ser transformada de modo a ser inserida em um modelo preditivo. Para mais detalhes sobre o passo de preparação a seguir, consulte o notebook de apresentação do módulo `ml.transformers`.

```python
# Importando classes
from pycomp.ml.transformers import FormataColunas, FiltraColunas, DefineTarget, \
                                   EliminaDuplicatas, SplitDados, DummiesEncoding
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Parâmetros de execução
initial_features = ['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']

# Construindo pipeline inicial
initial_pipeline = Pipeline([
    ('formatter', FormataColunas()),
    ('selector', FiltraColunas(features=initial_features)),
    ('target_prep', DefineTarget(target_col='survived', pos_class=1)),
    ('dup_dropper', EliminaDuplicatas()),
    ('splitter', SplitDados(target='target'))
])

# Executando pipeline inicial
X_train, X_test, y_train, y_test = initial_pipeline.fit_transform(df_train)

# Separando features por tipo primitivo
num_features = [col for col, dtype in X_train.dtypes.items() if dtype != 'object']
cat_features = [col for col, dtype in X_train.dtypes.items() if dtype == 'object']

# Construindo pipeline numérico
num_pipeline = Pipeline([
    ('na_filler', SimpleImputer(strategy='median'))
])

# Construindo pipeline categórico
cat_pipeline = Pipeline([
    ('encoder', DummiesEncoding(dummy_na=True))
])

# Construindo pipeline final de prep
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])

# Executando pipeline completo nos dados de treino e de teste
X_train_prep = full_pipeline.fit_transform(X_train)
X_test_prep = full_pipeline.fit_transform(X_test)

# Retornando features
pipeline_encoder = full_pipeline.named_transformers_['cat'].named_steps['encoder']
cat_encoded_features = pipeline_encoder.features_after_encoding
features = num_features + cat_encoded_features

# Verificando resultados
print(f'Dimensões da base de treino X_train_prep: {X_train_prep.shape}')
print(f'Dimensões da base de teste X_test_prep: {X_test_prep.shape}')
print(f'Quantidade de features consideradas: {len(features)}')
```

# Modelagem: Trainer

Após a construção e execução de um fluxo completo de preparação dos dados a partir de pipelines construídos com as ferramentas `pycomp`, é possível agora iniciar os passos para o treinamento e avaliação de modelos de classificação. Para usufruir das funcionalidades do módulo `ml.trainer`, é preciso antes definir quais modelos serão utilizados para treinamento e seus respectivos hiperparâmetros de busca.

## Definições Iniciais

Nessa sessão, serão realizadas algumas definições iniciais em termos de análise e construção de estruturas de requisito para utilização do módulo `pycomp.ml.trainer`.

```python
# Importando classificadores candidatos
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Instanciando modelos
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
forest = RandomForestClassifier()

# Definindo espaço de hiperparâmetros para cada modelo
# Logistic Regression
logreg_param_grid = {
    'C': np.linspace(0.1, 10, 20),
    'penalty': ['l1', 'l2'],
    'class_weight': ['balanced', None],
    'random_state': [42],
    'solver': ['liblinear']
}

# Decision Trees
dtree_param_grid = {
    'criterion': ['entropy', 'gini'],
    'max_depth': [3, 5, 10, 20],
    'max_features': np.arange(1, X_train_prep.shape[1]),
    'class_weight': ['balanced', None],
    'random_state': [42]
}

# Random Forest
forest_param_grid = {
    'bootstrap': [True, False],
    'max_depth': [3, 5, 10, 20, 50],
    'n_estimators': [50, 100, 200, 500],
    'random_state': [42],
    'max_features': ['auto', 'sqrt'],
    'class_weight': ['balanced', None]
}
```

Na sequência, será preparado um dicionário de classificadores que será passado como _input_ para a classe `ClassificadorBinario` do módulo `pycomp.ml.trainer`. Esse dicionário é essencial para que seja possível iterar por cada combinação de modelo e conjunto de hiperparâmetros e realizar os passos de treino e avaliação de forma automática.

```python
# Preparando set de classificadores
set_classifiers = {
    'LogisticRegression': {
        'model': logreg,
        'params': logreg_param_grid
    },
    'DecisionTrees': {
        'model': dtree,
        'params': dtree_param_grid
    },
    'RandomForest': {
        'model': forest,
        'params': forest_param_grid
    }
}
```

Por fim, é possível agora inserir esse dicionário de classificadores nos métodos de treinamento da classe `ClassificadorBinario`, permitindo assim a avaliação de todos os modelos de uma única vez.

## Treinamento de Modelos

Com a definição do dicionário de classificadores `set_classifiers`, é possível então utilizar as funcionalidades do `pycomp` para treinar os modelos selecionados.

```python
# Importando classe
from pycomp.ml.trainer import ClassificadorBinario

# Instanciando classe
trainer = ClassificadorBinario()

# Treinando modelos
trainer.training_flow(set_classifiers, X_train_prep, y_train, X_test_prep, y_test, features, 
                      random_search=True)
```

Basicamente, o método `training_flow()` realizou o treinamento de todos os modelos definidos no dicionário `set_classifiers` (aplicando `RandomizedSearchCV` conforme configurado pelo argumento `random_search` do método), aliou a performance de todos os modelos utilizando os dados de treino e teste e, por fim, computou a importância das features para cada um dos modelos, salvando assim esses resultados em um diretório de output (default=`output/`).

Em um desenvolvimento padrão, o cientista/analista de dados deveria realizar treinamentos individuais dos modelos, chamar métodos específicos do `sklearn` para aplicação de validação cruzada nos dados de treino, aplicar predições nos dados de testes, computar as demais métricas envolvidas (precision, recall, score, f1, auc), chamar o método feature_importances_ pra cada classificador e analisar individualmente esses resultados. Com o método `trainin_flow()`, todos esses passos são encapsulados e consolidados em um método único. 

Para ilustrar os resultados obtidos nessa etapa de treinamento, serão lidos os arquivos `metrics.csv` e `top_features.csv` gerados no diretório `output/`.

```python
# Verificando resultados do training_flow (metrics)
metrics = pd.read_csv('output/metrics/metrics.csv')
metrics
```

```python
# Verificando resultados do training_flow (top features)
top_features = pd.read_csv('output/metrics/top_features.csv')
top_features.head()
```

De maneira totalmente automática e encapsulada, o método `training_flow()` da classe `ClassificadorBinario` treinou, buscou pelas melhores combinações de hiperparêmtros e avaliou os classificadores disponibilizados no dicionário `set_classifiers` passado como input. O arquivo `metrics.csv` foi gerado de modo a evidenciar as principais métricas de classificação para os modelos utilizando dados de treino (validação cruzada) e teste.

Já o arquivo `top_features.csv` traz consigo uma análise das features mais importantes pra cada um dos classificadores treinados (desde que estes tenham o método `feature_importances_` disponível). Em outras palavras, foi possível ter em mãos uma rica análise dos modelos com apenas a execução de um único método da classe `ClassificadorBinario`.

## Análise Visual dos Modelos

Muito além de ter em mãos bases e arquivos contendo resultados analíticos dos classificadores, a classe `ClassificadorBinario` proporciona também a retirada de insights valiosos a partir de análises gráficas criadas com o objetivo de facilitar a comparação e seleção do melhor modelo de Machine Learning para a respectiva tarefa. Nos próximos tópicos, os métodos do objeto `trainer` instanciado serão executados de modo a mostrar as diversas possibilidades trazidas pelo pacote `pycomp`.

### Métricas

A partir do método `plot_metrics()` é possível analisar graficamente e comparar as diferentes métricas para os diferentes modelos treinados

```python
# Executando método
trainer.plot_metrics()
```

### Feature Importance

O método `plot_feature_importance()` realiza a plotagem das features mais importantes para cada classificador treinado (desde que este possua o método `feature_importances_` disponível por padrão)

```python
# Plotando features mais importantes
trainer.plot_feature_importance(features=features)
```

Na plotagem acima e nas mensagens de log retornadas pela classe, é possível perceber que o modelo `LogisticRegression` não possui o método `feature_importances_`.

### Matriz de Confusão

Entrando mais a fundo na análise dos classificadores treinados, o método `plot_confusion_matrix()` é responsável por plotar e customizar uma matriz de confusão para cada modelo utilizando predições obtidas com os dados de treino (validação cruzada) e de teste.

```python
# Visualizando matrizes de confusão
trainer.plot_confusion_matrix()
```

### Curva ROC

Continuando com a análise visual, o método `plot_roc_curve()` retorna alguns parâmetros relacionados a taxa de falsos positivos para a plotagem da famosa curva ROC para os modelos treinados (dados de treino e de teste).

```python
# Plotando curva ROC para os modelos
trainer.plot_roc_curve()
```

### Distribuição de Probabilidades

Uma outra análise extreamente interessante em modelos de classificação binária é a análise de densidade de probabilidade. Com `plot_score_distribution()` é responsável pela aplicação do método `predict_proba()` em cada um dos classificadores treinados, retornando assim a probabilidade de ocorrência da classe positiva para os elementos da base de dados individualmente, abrindo a possibilidade de plotar a densidade de cada score.

```python
# Plotando densidade de probabilidades pro score obtido
trainer.plot_score_distribution()
```

### Volume por Faixa de Score

Em complemento a análise de distribuição de probabilidade dos modelos, o método `plot_score_bins()` traz uma análise discreta por faixa de score (definida pelo argumento `bin_range`) pra cada um dos modelos treinados. Com isso, é possível ter uma ideia prática sobre a quantidade de amostras das classes positiva e negativa preditas pelos modelos em cada faixa de score definida.

Além do volume absoluto dos elementos, o método `plot_score_bins()` traz também um conjunto de plotagens em termos percentuais, permitindo assim uma análise de representatividade da classe por faixa de score.

```python
# Plotando volumetria de elementos por faixa de score
trainer.plot_score_bins(bin_range=.20)
```

### Curva de Aprendizado

A análise das chamadas curvas de aprendizado (ou `learning curves`) é uma forma extremamente rica de avaliar se os modelos treinados sofrem de _overfitting_ ou _underfitting_. Para isso, a base de dados é dividida em chuncks (ou mini conjuntos) utilizados para o treinamento e avaliação de forma sequencial na própria base de treino e através de validação cruzada. 

Dessa forma, é possível analisar, por exemplo, se o erro em validação cruzada é distante do erro na base pura de treino (indicativo de overfitting).

```python
# Plotando curva de aprendizado
trainer.plot_learning_curve()
```

### Análise Shap

Finalizando o conjunto de análises visuais do módulo `ml.trainer` do pacote `pycomp` (até a versão 0.0.16), é possível analisar, através da execução do método `plot_shap_analysis()` a famosa análise shap dos modelos através de um `summary_plot()` do tipo violino. Tal análise pode ser utilizada para interpretação dos modelos em termos das features envolvidas e em seus respectivos valores e impactos no output.

Para sua execução, basta passar um modelo alvo da análise (argumento `model_name`) e o conjunto de features utilizado no treinamento (argumento `features`)

```python
# Plotando análise shap
trainer.plot_shap_analysis(model_name='RandomForest', features=features)
```

É importante ressaltar que existe ainda um método da classe nomeado como `visual_analysis()` que, por sua vez, é responsável por encapsular a execução sequencial de cada um dos métodos gráficos mostrados acima, permitindo assim uma forma automática de realizar todas as análises visuais em uma única chamada.

A cada execução dos métodos de análises gráficas é gerada uma saída em formato _.png_ em um diretório de output pré-defindo (`outputs/img`). Ao final do processo de modelagem, além das plotagens gráficas no próprio notebook utilizado para o desenvolvimento, é esperado que as imagens e os arquivos gerados pelo processo estejam salvos no diretório padrão de output (ou então por algum outro diretório definido pelo argumento `output_path` definido em cada método da classe `ClassificadorBinario`).
