# Leitura dos Dados

Observando os [metadadados](https://www.kaggle.com/sakshigoyal7/credit-card-customers) do dataset disponíveis no Kaggle, é possível detalhar cada uma das 23 colunas da base como:

- **_CLIENTNUM:_** identificador único do cliente cartonista;
- **_Attrition_Flag:_** evento interno relacionado a atividade do cliente;
- **_Customer_Age:_** idade do cliente (em anos);
- **_Gender:_** gênero do cliente (M=Masculino, F=Feminino);
- **_Dependent_count:_** número de dependentes do cliente;
- **_Education_Level:_** nível escolar do cliente;
- **_Marital_Status:_** estado civil do cliente;
- **_Income_Category:_** categoria relacionada ao salário anual do cliente;
- **_Card_Category:_** categoria do cartão de crédito (Blue, Silver, Gold ou Platinum);
- **_Months_on_book:_** período de relacionamento com o banco (em meses)

___

_Os demais atributos presentes na base não possuem descrições detalhadas na página de metadados. Entretanto, de forma intuitiva, é possível extrair algum significado destes a partir do nome das colunas registrados_
___

- **_Total_Relationship_Count:_** indicador de relacionamento geral do cliente com o banco;
- **_Months_Inactive_12_mon:_** quantidade de meses de inatividade do cliente considerando os últimos 12 meses;
- **_Contacts_Count_12_mon:_** quantidade de contatos registrados pelo cliente considerando os últimos 12 meses;
- **_Credit_Limit:_** limite de crédito do cliente;
- **_Total_Revolving_Bal:_**
- **_Avg_Open_To_Buy:_** indicador de disposição de compra por parte do cliente (abertura de oferta);
- **_Total_Amt_Chng_Q4_Q1:_** provavelmente indica o valor migrado entre Q4 e Q1 de um período anual cheio;
- **_Total_Trans_Amt:_** total transacionado pelo cliente;
- **_Total_Trans_Ct:_** total transacionado pelo cliente no cartão;
- **_Total_Ct_Chng_Q4_Q1:_** provavelmente indica o valor migrado de transações no cartão entre Q4 e Q1 de um período anual cheio;
- **_Avg_Utilization_Ratio:_** indicador de utilização média do cartão por parte do cliente;

___
_Os dois últimos atributos presentes na base indicam, de alguma forma, resultados de processos de classificação e construção de score para os clientes_

**_Análise de correlação entre variáveis_**
___

Após uma abordagem inicial em variáveis importantes que descrevem o perfil dos clientes da instituição financeira, é possível analisar os atributos numéricos de uma única vez em termos de correlação. Essa abordagem é importante para dar uma visão geral, em uma única visão, sobre como as variáveis estão correlacionadas entre si e com uma variável target (`Attrition_flag`)

```python
# Matriz de correlação
tmp = df.copy()
tmp['churned'] = tmp['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
clf_drop_cols = ['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
                 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
plot_corr_matrix(df=tmp.drop(clf_drop_cols, axis=1), corr_col='churned', figsize=(13, 13))
```

A partir da matriz de correlação acima, é possível analisar os principais fatores que possivelmente influenciam no _churn_ de clientes desta instituição financeira. A função `plot_corr_matrix()` do módulo `pycomp` analisa a correlação (default=`Pearson`) das variáveis numéricas com uma variável target (argumento `corr_col`) e também entre si, permitindo assim um detalhamento na análise das principais variáveis correlacionadas. Dessa forma, é possível citar:

1. As variáveis `Total_Trans_Ct`, `Total_Ct_Chng_Q4_Q1` e `Total_Revolving_Bal` são as top 3 features que mais influenciam diretamente e positivamente o _churn_ de clientes. Em outras palavras, quanto maior o valor dessas 3 variáveis mencionadas, mais alto o rate de _churn_ desses clientes.
2. No outro espectro de análise, as variáveis `Contacts_Count_12_mon` e `Months_Inactive_12_mon` são as 2 principais features que apresentam correlação negativa com a variável target de churn. Isso significa que, quanto menor o valor dessas 2 variávies mencionadas, mais alto o rate de _churn_ do público.
3. Analisando as correlações das variáveis entre si, é possível citar:
    * As variáveis `Total_Trans_Amt` e `Total_Trans_Ct` possuem um alto índice de correlação positiva (crescimentos diretamente proporcionais) - quanto maior o valor total transacionado, maior o valor total transacionado no cartão
    * As variáveis `Avg_Utilization_Ratio` e `Total_Revolving_Bal` possuem um alto índice de correlação positiva (crescimentos diretamente proporcionais)
    * As variáveis `Customer_Age` e `Months_on_book` possuem um alto índice de correlação positiva (crescimentos diretamente proporcionais) - quanto maior a idade do cliente, maior o tempo de relacionamento dele com o banco
    * Existe uma relação inversamente proporcional entre as variáveis `Credit_Limit` e `Avg_Utilization_Ratio`, indicando que quanto menor a média de uso do cliente aos produtos, menor seu limite pré-aprovado;
    * Essa mesma relação inversa também ocorre entre `Avg_Open_To_Buy` e `Avg_Utilization_Ratio`, indicando que um menor uso também influencia na abertura de compra do cliente.
4. O limite pré-aprovado não influencia no _churn_

```python
plot_distplot(df=df, col='Total_Trans_Ct', hue='Attrition_Flag', kind='kde',
              title='Distribuição de volume transacionado no cartão\npor flag de atrito de clientes')
```

Analisando as correlações e propondo uma visão mais detalhada das variáveis com maior impacto em clientes que deixaram o banco, é possível perceber que clientes atritados possuem uma distribuição menor de volume transacionado no cartão, indicando assim uma possível insatisfação com os serviços e uma possível migração para outras instituições.

```python
plot_distplot(df=df, col='Total_Ct_Chng_Q4_Q1', hue='Attrition_Flag', kind='strip', 
              palette=['cadetblue', 'crimson'],
              title='Distribuição da variação de transações do Q4 para o Q1\npor flag de atrito de clientes')
```

## Pipeline Inicial

Nessa sessão, serão propostas transformações na base de dados comuns a todo o conjunto. O grande objetivo em criar um pipeline inicial de preparação de dados é garantir que algumas etapas sejam aplicadas antes de um pipeline oficial entrar em cena, como por exemplo, um `drop` inicial nas features ou a remoção de dados duplicadas de uma base de treino.

### Features Candidatas

Os objetivos dessa etapa são:
1. Filtrar as colunas iniciais a serem utilizadas na modelagem (eliminação de colunas de chave ou não representativas para um modelo preditivo)
2. Preparar um transformador que possa realizar esse processo de forma automática, caso haja a necessidade de repetir todo o processo de treinamento

```python
# Importando classe
from pycomp.ml.transformers import FiltraColunas

# Definições iniciais
TO_DROP = ['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
           'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
INITIAL_FEATURES = list(df.drop(TO_DROP, axis=1).columns)

# Criando e aplicando transformador de seleção de features
selector = FiltraColunas(features=INITIAL_FEATURES)
df_slct = selector.fit_transform(df)

# Resultados
print(f'Dimensões da base original: {df.shape}')
print(f'Dimensões da base após feature selection: {df_slct.shape}')
```

### Definição do Target

Na nossa base de dados original, a coluna que identifica o target do problema de negócio é dada por `Attrition_Flag`. Para que seja possível treinar um modelo preditivo, é preciso realizar uma preparação nessa coluna para transformá-la em "zeros e ums". Para isso, é possível utilizar uma classe `DefineTarget` que, por sua vez, é responsável por aplicar uma modificação em uma base de dados baseado em uma coluna target e em uma entrada dada como classe positiva.

```python
# Importando classe
from pycomp.ml.transformers import DefineTarget

# Criando objeto e aplicando transformação
target_transformer = DefineTarget(target_col='Attrition_Flag', pos_class='Attrited Customer')
df_tgt = target_transformer.fit_transform(df_slct)

df_tgt['target'].value_counts()
```

### Dados de Treino e de Teste

Os objetivos dessa sessão são:

1. Definir a coluna target do modelo;
2. Aplicar separação de dados em treino e teste

```python
# Importando classe
from pycomp.ml.transformers import SplitDados

# Criando objeto e aplicando transformador
TARGET = 'target'
splitter = SplitDados(target=TARGET)
X_train, X_test, y_train, y_test = splitter.fit_transform(df_slct)

# Resultados
print(f'Dimensões de X_train: {X_train.shape}')
print(f'Dimensões de X_test: {X_test.shape}')
print(f'Dimensões de y_train: {y_train.shape}')
print(f'Dimensões de y_test: {y_test.shape}')
```

### Transformação Logaritma

Para validar o impacto da transformação logaritma nos modelos preditivos candidatos, opcionalmente vamos propor uma etapa no pipeline que aplique esse procedimento nas features numéricas presentes na base. Com isso, podemos validar se a performance final do modelo é sensível a esse tipo de transformação.

```python
# Exemplo de distribuição
log_ex = 'Credit_Limit'
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(17, 7))
plot_distplot(df=X_train_num, col='Credit_Limit', ax=axs[0], hist=True,
              title=f'Original {log_ex} Distribution')

tmp_data = X_train_num.copy()
tmp_data['Credit_Limit'] = tmp_data['Credit_Limit'].apply(lambda x: np.log1p(x))
plot_distplot(df=tmp_data, col='Credit_Limit', ax=axs[1], color='mediumseagreen', hist=True, 
              title=f'{log_ex} After Log Transformation')
```

Duas medidas estatísticas altamente relevantes para análise de distribuição são `skew` e `kurtosis`. Através do [link](https://codeburst.io/2-important-statistics-terms-you-need-to-know-in-data-science-skewness-and-kurtosis-388fef94eeaa) é possível ter uma ideia clara sobre o que é cada uma dessas medidas e como interpretar distribuições contínuas através de seus valores. 

A transformação logaritma auxilia no aumento da performance para distribuições com `skewness` positiva (assimétricas à esquerda). Assim, vamos analisar novamente as features numéricas e rankear as principais features com oportunidade de melhoria através desse tipo de transformação.

```python
from scipy.stats import skew, kurtosis

tmp_ov = df_overview.copy()
tmp_ov['skew'] = tmp_ov.query('feature in @num_features')['feature'].apply(lambda x: skew(X_train_num[x]))
tmp_ov['kurtosis'] = tmp_ov.query('feature in @num_features')['feature'].apply(lambda x: kurtosis(X_train_num[x]))
tmp_ov[~tmp_ov['skew'].isnull()].sort_values(by='skew', ascending=False).loc[:, ['feature', 'skew', 'kurtosis']]
```

A tabela acima mostra uma relação das features através de suas medidas de simetria `skewness` e `kurtosis`. No bloco de código abaixo, iremos executar a classe `DynamicLogTransformation` que, por sua vez, tem o papel de aplicar a transformação logaritma em uma base de dados em um pipeline de preparação. A vantagem dessa classe é a definição prévia de uma lista de features as quais a transformação será aplicada, sendo esta definida pelo usuário.

```python
# Importando classe
from pycomp.ml.transformers import DynamicLogTransformation

# Definindo parâmetros
COLS_TO_LOG = ['Total_Trans_Amt', 'Credit_Limit', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1',
               'Avg_Utilization_Ratio']
log_tr = DynamicLogTransformation(num_features=num_features, cols_to_log=COLS_TO_LOG)
X_train_num_ori = X_train_num.copy()
X_train_num_log = log_tr.fit_transform(X_train_num)

# Verificando alguns resultados
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(17, 12))

for i in range(3):
    ax = axs[0, i]
    plot_distplot(df=df, col=COLS_TO_LOG[i], ax=ax, hist=True,
                  title=f'{COLS_TO_LOG[i]} Distribution $Before$ \nLog Transformation')
    
for i in range(3):
    ax = axs[1, i]
    plot_distplot(df=X_train_num_log, col=COLS_TO_LOG[i], ax=ax, hist=True, color='mediumseagreen',
                  title=f'{COLS_TO_LOG[i]} Distribution $After$ \nLog Transformation')

plt.tight_layout()
```

Adicionalmente, é válido citar que a class e `DynamicLogTransformation` possui um atributo booleano chamado `application` que pode ser utilizado futuramente para interações em `GridSearch` ou `RandomizedSearch`. Seu objetivo é possibilitar análises em performance de modelos **com** ou **sem** a transformação logaritma.

## Pipelines Completos

Uma vez definidos os passos a serem considerados na preparação dos dados, seja esta inicial ou oficial, temos a capacidade agora de consolidar todos os _steps_ em blocos únicos transformação de dados. Assim, a célula abaixo tem por objetivo realizar esse processo de consolidação ao mesmo tempo que define algumas variáveis globais de projeto.

```python
# Importando bibliotecas
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Definindo variáveis globais
ORIGINAL_TARGET = 'Attrition_Flag'
TARGET = 'target'
TARGET_POSITIVE_CLASS = 'Attrited Customer'

INITIAL_FEATURES = ['Customer_Age', 'Gender', 'Dependent_count', 'Education_Level', 'Marital_Status',
                    'Income_Category', 'Card_Category', 'Months_on_book', 'Total_Relationship_Count',
                    'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Attrition_Flag',
                    'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt',
                    'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']

INITIAL_PRED_FEATURES = [col for col in INITIAL_FEATURES if col not in [ORIGINAL_TARGET, TARGET]]

NUM_FEATURES = ['Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
                'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
                'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct',
                'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']

CAT_FEATURES = ['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']

MODEL_FEATURES = ['Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
                  'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
                  'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct',
                  'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio', 'Gender_F', 'Gender_M',
                  'Education_Level_College', 'Education_Level_Doctorate', 'Education_Level_Graduate',
                  'Education_Level_High School', 'Education_Level_Post-Graduate',
                  'Education_Level_Uneducated', 'Education_Level_Unknown', 'Marital_Status_Divorced',
                  'Marital_Status_Married', 'Marital_Status_Single', 'Marital_Status_Unknown',
                  'Income_Category_$120K +', 'Income_Category_$40K - $60K', 'Income_Category_$60K - $80K',
                  'Income_Category_$80K - $120K', 'Income_Category_Less than $40K',
                  'Income_Category_Unknown', 'Card_Category_Blue', 'Card_Category_Gold',
                  'Card_Category_Platinum', 'Card_Category_Silver']

SCALER_TYPE = 'Standard'
ENCODER_DUMMY_NA = False
LOG_APPLICATION = True
COLS_TO_LOG = ['Total_Trans_Amt', 'Credit_Limit', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1',
               'Avg_Utilization_Ratio']

# Construindo pipeline inicial (treino e predição)
initial_train_pipeline = Pipeline([
    ('col_filter', FiltraColunas(features=INITIAL_FEATURES)),
    ('target_transformer', DefineTarget(target_col=ORIGINAL_TARGET, pos_class=TARGET_POSITIVE_CLASS))
])

initial_pred_pipeline = Pipeline([
    ('col_filter', FiltraColunas(features=INITIAL_PRED_FEATURES))
])

# Construindo pipeline numérico
num_pipeline = Pipeline([
    ('log_transformer', DynamicLogTransformation(application=LOG_APPLICATION, num_features=NUM_FEATURES, 
                                                 cols_to_log=COLS_TO_LOG)),
    ('scaler', DynamicScaler(scaler_type=SCALER_TYPE))
])

# Construindo pipeline categórico
cat_pipeline = Pipeline([
    ('encoder', DummiesEncoding(dummy_na=ENCODER_DUMMY_NA))
])

# Criando pipeline completo
prep_pipeline = ColumnTransformer([
    ('num', num_pipeline, NUM_FEATURES),
    ('cat', cat_pipeline, CAT_FEATURES)
])
```

```python
# Lendo base bruta de dados
df = pd.read_csv(os.path.join(DATA_PATH, FILENAME))

# Executando pipeline inicial de treino
df_prep = initial_train_pipeline.fit_transform(df)

# Separando dados de treino e de teste
X_train, X_test, y_train, y_test = train_test_split(df_prep.drop(TARGET, axis=1), df_prep[TARGET].values,
                                                    test_size=.20, random_state=42)

# Executando pipelines de preparação
X_train_prep = prep_pipeline.fit_transform(X_train)
X_test_prep = prep_pipeline.fit_transform(X_test)

# Validando dimensões
print(f'Dimensões de X_train_prep: {X_train_prep.shape}')
print(f'Dimensões de X_test_prep: {X_test_prep.shape}')
print(f'\nQuantidade de features consideradas: {len(MODEL_FEATURES)}')
```

# Modelagem: Prevendo Churn de Clientes

Por fim, após extensas etapas de análise exploratória e preparação da base de dados, é chegada a hora de aplicar conceitos de Machine Learning para o desenvolvimento de um modelo preditivo capaz de prever a perda ou a migração de clientes para outras instituições bancárias. Em posse da base final preparada, vamos propor alguns algoritmos capazes de nos dar essa resposta e, através do treinamento e avaliação dos mesmos, vamos escolher o melhor modelo para a tarefa em questão.

Todos esses passos serão construídos com base nas ferramentas disponíveis no pacote `pycomp` através de seu módulo `pycomp.ml.trainer` (mais especificamente na classe `ClassificadorBinario`). ## Definições Iniciais

Nessa etapa, serão definidas algumas variáveis importantes para a utilização da classe `ClassificadorBinario` do pacote `pycomp`. É neste momento que definimos as estruturas e os objetos que servirão de input para o treinamento e avaliação dos modelos.

```python
# Importando modelos
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

# Instanciando objetos
dtree = DecisionTreeClassifier()
forest = RandomForestClassifier()
lgbm = LGBMClassifier()
xgb = XGBClassifier()

# Criando dicionário set_classifiers
model_obj = [dtree, forest, lgbm, xgb]
model_names = [type(model).__name__ for model in model_obj]
set_classifiers = {name: {'model': obj, 'params': {}} for (name, obj) in zip(model_names, model_obj)}

print(f'Classifiers that will be trained on next steps: \n\n{model_names}')
```

## Treinando Modelos

Uma vez preparada a estrutura de modelagem a partir de objetos específicos, como por exemplo, o dicionário `set_classifiers`, é possível agora importar a classe `ClassificadorBinario` presente no módulo `pycomp.ml.trainer` para realizar todo o treinamento e avaliação dos modelos candidatos.

Essa classe foi desenvolvida com o intuito de facilitar grandemente o trabalho do analista/cientista em termos de implementação de códigos para treinar, avaliar e otimizar modelos preditivos de classificação binária. Seus métodos comportam poderosas funcionalidades que realizam diversas ações com apenas uma chamada.

```python
# Importando classe
from pycomp.ml.trainer import ClassificadorBinario

# Criando objeto e realizando treinamento
trainer = ClassificadorBinario()
trainer.fit(set_classifiers, X_train_prep, y_train, random_search=False)
```

O método `fit()` do objeto `trainer` criado tem a responsabilidade de realizar o treinamento dos modelos encapsulados no dicionário `set_classifiers` criado na etapa de definições iniciais.

Configurando o método para aplicar também o processo de `RandomizedSearchCV` (busca aleatória dos melhores hiperparâmetros de cada algoritmo), é possível construir modelos otimizados de acordo com o espaço de busca passado no dicionário `set_classifiers`.

## Avaliando Performance

Uma vez treinados os modelos candidatos através do método `fit()`, é possível então avaliar a performance obtida em cada um dos casos, retornando assim as principais métricas de classificação capazes de nos indicar o melhor direcionamento para a dada tarefa.

Para realizar esse processo, podemos utilizar os métodos `evaluate_performance()` ou `plot_metrics()` do objeto `trainer`. Como mencionado anteriormente, o método `evaluate_performance()` retorna uma tabela analítica contendo a performance de cada modelo (em treino e teste) para as principais métricas de avaliação de modelos de classificação. ## Fine Tuning

Vimos, nos tópicos acima, a boa performance obtida pelo modelo `LightGBM`. Nessa sessão, vamos utilizar este algoritmo para aplicar uma tunagem de hyperparâmetros em busca de uma melhora ainda maior na performance final.

```python
# Importando bibliotecas
from sklearn.model_selection import RandomizedSearchCV
from pycomp.ml.transformers import SeletorTopFeatures
from pycomp.ml.trainer import clf_cv_performance

# Coletando modelo
model_name = 'LGBMClassifier'
baseline_model = trainer._get_estimator(model_name)
feature_importance = baseline_model.feature_importances_

# Retornando base inteira para treinamento (treino + teste)
X = X_train.append(X_test)
y = np.concatenate((y_train, y_test))

# Criando pipeline de preparação e predição
tunning_pipeline = Pipeline([
    ('prep', prep_pipeline),
    ('selector', SeletorTopFeatures(feature_importance, k=len(MODEL_FEATURES))),
    ('model', baseline_model)
])

# Definindo hiperparâmetros de busca
tunning_params_grid = [{
    'prep__num__log_transformer__application': [True, False],
    'prep__num__scaler__scaler_type': [None, 'Standard', 'MinMax'],
    'selector__k': np.arange(1, len(MODEL_FEATURES) + 1, 1),
    'model__boosting_type': ['gbdt', 'dart', 'goss'],
    'model__class_weight': [None, 'balanced'],
    'model__colsample_bytree': np.linspace(.5, 1, 50),
    'model__importance_type': ['split', 'gain'],
    'model__learning_rate': np.linspace(1e-5, 1, 300),
    'model__max_depth': np.arange(-1, 100, 2),
    'model__min_child_samples': np.arange(10, 50, 1),
    'model__min_child_weight': np.linspace(1e-4, 1, 100),
    'model__n_estimators': [50, 100, 150, 200, 300, 500],
    'model__num_leaves': np.arange(5, 100, 1),
    'model__objective': ['binary'],
    'model__random_state': [42],
    'model__reg_alpha': np.linspace(.0, 1.0, 50)
}]

# Criando busca
tunning_search = RandomizedSearchCV(tunning_pipeline, tunning_params_grid, scoring='accuracy', cv=5,
                                    n_jobs=-1, verbose=-1, random_state=42)
tunning_search.fit(X, y)
print(f'Best hyperparameters found by RandomizedSearch:\n')
for k, v in tunning_search.best_params_.items():
    print(f'{k}: {v}')
```

```python
# Final metrics
final_pipeline = tunning_search.best_estimator_
final_metrics = clf_cv_performance(final_pipeline, X, y)
final_metrics
```
