# Leitura dos Dados

Após uma definição formal do contexto do projeto, o alinhamento dos objetivos deste trabalho e, por fim, a definição das variáveis do projeto, é possível iniciar as investigações a partir da leitura da base de dados disponível para análise.

```python
# Lendo dados e visualizando base
df = pd.read_csv(os.path.join(DATA_PATH, FILENAME))
print(f'Dimensões da base de dados: {df.shape}')
df.head()
```

Observando os [metadadados](https://www.kaggle.com/sakshigoyal7/credit-card-customers) do dataset disponíveis no Kaggle, é possível detalhar cada uma das 23 colunas da base como:

- **_CLIENTNUM:_** identificador único do cliente cartonista;
- **_Attrition_Flag:_** evento interno relacionado a atividade do cliente;
- **_Customer_Age:_** idade do cliente (em anos);
- **_Gender:_** gênero do cliente (M=Masculino, F=Feminino);
- **_Dependent_count:_** número de dependentes do cliente;
- **_Education_Level:_** nível escolar do cliente;
- **_Marital_Status:_** estado civil do cliente;
- **_Income_Category:_** categoria relacionada ao salário anual do cliente;
- **_Card_Category:_** categoria do cartão de crédito (Blue, Silver, Gold ou Platinum);
- **_Months_on_book:_** período de relacionamento com o banco (em meses)

___

_Os demais atributos presentes na base não possuem descrições detalhadas na página de metadados. Entretanto, de forma intuitiva, é possível extrair algum significado destes a partir do nome das colunas registrados_
___

- **_Total_Relationship_Count:_** indicador de relacionamento geral do cliente com o banco;
- **_Months_Inactive_12_mon:_** quantidade de meses de inatividade do cliente considerando os últimos 12 meses;
- **_Contacts_Count_12_mon:_** quantidade de contatos registrados pelo cliente considerando os últimos 12 meses;
- **_Credit_Limit:_** limite de crédito do cliente;
- **_Total_Revolving_Bal:_**
- **_Avg_Open_To_Buy:_** indicador de disposição de compra por parte do cliente (abertura de oferta);
- **_Total_Amt_Chng_Q4_Q1:_** provavelmente indica o valor migrado entre Q4 e Q1 de um período anual cheio;
- **_Total_Trans_Amt:_** total transacionado pelo cliente;
- **_Total_Trans_Ct:_** total transacionado pelo cliente no cartão;
- **_Total_Ct_Chng_Q4_Q1:_** provavelmente indica o valor migrado de transações no cartão entre Q4 e Q1 de um período anual cheio;
- **_Avg_Utilization_Ratio:_** indicador de utilização média do cartão por parte do cliente;

___
_Os dois últimos atributos presentes na base indicam, de alguma forma, resultados de processos de classificação e construção de score para os clientes_
___

**_Análise de correlação entre variáveis_**
___

Após uma abordagem inicial em variáveis importantes que descrevem o perfil dos clientes da instituição financeira, é possível analisar os atributos numéricos de uma única vez em termos de correlação. Essa abordagem é importante para dar uma visão geral, em uma única visão, sobre como as variáveis estão correlacionadas entre si e com uma variável target (`Attrition_flag`)

```python
# Matriz de correlação
df['churned'] = df['Attrition_Flag'].map({'Existing Customer': 1, 'Attrited Customer': 0})
clf_drop_cols = ['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
                 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
plot_corr_matrix(df=df.drop(clf_drop_cols, axis=1), corr_col='churned', figsize=(13, 13))
```

A partir da matriz de correlação acima, é possível analisar os principais fatores que possivelmente influenciam no _churn_ de clientes desta instituição financeira. A função `plot_corr_matrix()` do módulo `pycomp` analisa a correlação (default=`Pearson`) das variáveis numéricas com uma variável target (argumento `corr_col`) e também entre si, permitindo assim um detalhamento na análise das principais variáveis correlacionadas. Dessa forma, é possível citar:

1. As variáveis `Total_Trans_Ct`, `Total_Ct_Chng_Q4_Q1` e `Total_Revolving_Bal` são as top 3 features que mais influenciam diretamente e positivamente o _churn_ de clientes. Em outras palavras, quanto maior o valor dessas 3 variáveis mencionadas, mais alto o rate de _churn_ desses clientes.
2. No outro espectro de análise, as variáveis `Contacts_Count_12_mon` e `Months_Inactive_12_mon` são as 2 principais features que apresentam correlação negativa com a variável target de churn. Isso significa que, quanto menor o valor dessas 2 variávies mencionadas, mais alto o rate de _churn_ do público.
3. Analisando as correlações das variáveis entre si, é possível citar:
    * As variáveis `Total_Trans_Amt` e `Total_Trans_Ct` possuem um alto índice de correlação positiva (crescimentos diretamente proporcionais) - quanto maior o valor total transacionado, maior o valor total transacionado no cartão
    * As variáveis `Avg_Utilization_Ratio` e `Total_Revolving_Bal` possuem um alto índice de correlação positiva (crescimentos diretamente proporcionais)
    * As variáveis `Customer_Age` e `Months_on_book` possuem um alto índice de correlação positiva (crescimentos diretamente proporcionais) - quanto maior a idade do cliente, maior o tempo de relacionamento dele com o banco
    * Existe uma relação inversamente proporcional entre as variáveis `Credit_Limit` e `Avg_Utilization_Ratio`, indicando que quanto menor a média de uso do cliente aos produtos, menor seu limite pré-aprovado;
    * Essa mesma relação inversa também ocorre entre `Avg_Open_To_Buy` e `Avg_Utilization_Ratio`, indicando que um menor uso também influencia na abertura de compra do cliente.
4. O limite pré-aprovado não influencia no _churn_

```python
plot_distplot(df=df, col='Total_Trans_Ct', hue='Attrition_Flag', kind='kde',
              title='Distribuição de volume transacionado no cartão\npor flag de atrito de clientes')
```

Analisando as correlações e propondo uma visão mais detalhada das variáveis com maior impacto em clientes que deixaram o banco, é possível perceber que clientes atritados possuem uma distribuição menor de volume transacionado no cartão, indicando assim uma possível insatisfação com os serviços e uma possível migração para outras instituições.

## Pipelines Finais

Finalizando a etapa de Prep da análise preditiva de _churn_, vamos construir pipelines definitivos capazes de encapsular e executar todos os passos de tranformação detalhados nas sessões acima. Dessa forma, iremos:

1. Construir um pipeline inicial para preparação comum de uma base bruta recebida;
2. Construir um pipeline categórico para aplicação do processo de encoder;
3. Construir um pipeline único para unir o resultado do encoder à base de features numéricas;

Em termos de aplicação, iremos:

4. Aplicar pipeline inicial na base;
5. Separar os dados em treino e teste;
6. Aplicar separadamente o pipeline único nos dados de treino e teste.

```python
# Construindo classe para simplesmente passar os dados numéricos da base
from sklearn.base import BaseEstimator, TransformerMixin

class FakeNumPipe(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return X
```

```python
# Importando bibliotecas
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Construindo pipeline inicial
initial_pipeline = Pipeline([
    ('selector', FiltraColunas(features=INITIAL_FEATURES)),
    ('target_prep', DefineTarget(target_col='Attrition_Flag', pos_class='Attrited Customer')),
    ('splitter', SplitDados(target='target'))
])

# Construindo pipeline categórico
cat_pipeline = Pipeline([
    ('encoder', DummiesEncoding(dummy_na=False, cat_features_ori=cat_features))
])

# Construindo pipeline numérico
num_pipeline = Pipeline([
    ('fake', FakeNumPipe())
])

# Construindo pipeline único de prep
prep_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])
```

```python
# Aplicando pipeline inicial
X_train, X_test, y_train, y_test = initial_pipeline.fit_transform(df)

# Resultados
print(f'Dimensões de X_train: {X_train.shape}')
print(f'Dimensões de X_test: {X_test.shape}')
print(f'Dimensões de y_train: {y_train.shape}')
print(f'Dimensões de y_test: {y_test.shape}')
```

```python
# Aplicando pipeline completo
X_train_prep = prep_pipeline.fit_transform(X_train)
X_test_prep = prep_pipeline.fit_transform(X_test)

# Resultados
print(f'Dimensões de X_train: {X_train.shape}')
print(f'Dimensões de X_train_prep: {X_train_prep.shape}')
```

```python
# Retornando set de features finais do modelo
cat_encoded_features = prep_pipeline.named_transformers_['cat'].named_steps['encoder'].features_after_encoding
MODEL_FEATURES = num_features + cat_encoded_features
print(f'Total de features consideradas para o modelo: {len(MODEL_FEATURES)}')
```

Well done! Finalizamos a etapa de preparação da base com 2 objetos de pipelines sensacionais:

* **initial_pipeline:** responsável por receber uma base bruta de treino e aplicar os processos de:
    - Feature selection, selecionando apenas as features presentes no treinamento;
    - Preparação do target, transformando uma coluna target do modelo em 1s e 0s;
    - Separação treino/teste, permitindo um posterior treinamento e avaliação usando bases diferentes.
    
    
* **prep_pipeline:** responsável por receber uma base de saída do pipeline inicial e aplicar:
    - Encoding nos dados, permitindo a leitura e utilização de variáveis categóricas no modelo final

# Modelagem: Prevendo Churn

Uma vez finalizadas as etapas referentes a preparação da base de dados, finalmente podemos iniciar os passos de modelagem para treinamento de um modelo preditivo capaz de prever a probabilidade de _churn_ de clientes, ou seja, a probabilidade que cada cliente tem de encerrar suas atividades com a instituição financeira.

Assim, nessa sessão de modelagem, iremos realizar o treinamento de modelos preditivos (mais especificamente de modelos de classificação binária) em busca de selecionar o melhor algoritmo capaz de prever a probabilidade de churn dos clientes frente as principais métricas de análise.

## Definições Iniciais

Nessa sessão, serão realizadas algumas definições iniciais em termos de análise e construção de estruturas de requisito para utilização do módulo `pycomp.ml.trainer`.

```python
# Importando classificadores candidatos
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Instanciando modelos
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
forest = RandomForestClassifier()

# Definindo espaço de hiperparâmetros para cada modelo
# Logistic Regression
logreg_param_grid = {
    'C': np.linspace(0.1, 10, 20),
    'penalty': ['l1', 'l2'],
    'class_weight': ['balanced', None],
    'random_state': [42],
    'solver': ['liblinear']
}

# Decision Trees
dtree_param_grid = {
    'criterion': ['entropy', 'gini'],
    'max_depth': [3, 5, 10, 20],
    'max_features': np.arange(1, X_train_prep.shape[1]),
    'class_weight': ['balanced', None],
    'random_state': [42]
}

# Random Forest
forest_param_grid = {
    'bootstrap': [True, False],
    'max_depth': [3, 5, 10, 20, 50],
    'n_estimators': [50, 100, 200, 500],
    'random_state': [42],
    'max_features': ['auto', 'sqrt'],
    'class_weight': ['balanced', None]
}
```

Na sequência, será preparado um dicionário de classificadores que será passado como _input_ para a classe `ClassificadorBinario` do módulo `pycomp.ml.trainer`. Esse dicionário é essencial para que seja possível iterar por cada combinação de modelo e conjunto de hiperparâmetros e realizar os passos de treino e avaliação de forma automática.

```python
# Preparando set de classificadores
set_classifiers = {
    'LogisticRegression': {
        'model': logreg,
        'params': logreg_param_grid
    },
    'DecisionTrees': {
        'model': dtree,
        'params': dtree_param_grid
    },
    'RandomForest': {
        'model': forest,
        'params': forest_param_grid
    }
}
```

Por fim, é possível agora inserir esse dicionário de classificadores nos métodos de treinamento da classe `ClassificadorBinario`, permitindo assim a avaliação de todos os modelos de uma única vez.

## Treinamento de Modelos

Com a definição do dicionário de classificadores `set_classifiers`, é possível então utilizar as funcionalidades do `pycomp` para treinar os modelos selecionados.

```python
# Importando classe
from pycomp.ml.trainer import ClassificadorBinario

# Criando objeto e realizando treinamento
trainer = ClassificadorBinario()
trainer.fit(set_classifiers, X_train_prep, y_train, random_search=True, scoring='accuracy', cv=5, 
            verbose=-1, n_jobs=-1)
```

Basicamente, o método `fit()` realizou o treinamento de todos os modelos definidos no dicionário `set_classifiers` (aplicando `RandomizedSearchCV` conforme configurado pelo argumento `random_search` do método).

Em um desenvolvimento padrão, o cientista/analista de dados deveria realizar treinamentos individuais dos modelos, chamar métodos específicos do `sklearn` para aplicação de validação cruzada nos dados de treino, aplicar predições nos dados de testes, computar as demais métricas envolvidas (precision, recall, score, f1, auc), chamar o método feature_importances_ pra cada classificador e analisar individualmente esses resultados. Com o pacote `pycomp`, todos esses passos são encapsulados e consolidados em poucos métodos. 

```python
# Verificando resultados do treinamento (metrics)
metrics = trainer.evaluate_performance(X_train_prep, y_train, X_test_prep, y_test)
```

De maneira totalmente automática e encapsulada, o método `training_flow()` da classe `ClassificadorBinario` treinou, buscou pelas melhores combinações de hiperparêmtros e avaliou os classificadores disponibilizados no dicionário `set_classifiers` passado como input. O arquivo `metrics.csv` foi gerado de modo a evidenciar as principais métricas de classificação para os modelos utilizando dados de treino (validação cruzada) e teste.

Já o arquivo `top_features.csv` traz consigo uma análise das features mais importantes pra cada um dos classificadores treinados (desde que estes tenham o método `feature_importances_` disponível).
