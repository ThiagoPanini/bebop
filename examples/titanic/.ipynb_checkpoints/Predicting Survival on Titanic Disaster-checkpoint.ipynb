# Reading Data

After importing the project libraries and defining some important variables to be used along the development, now it's possible to read and understand the data available for the Titanic machine learning task.

```python
# Reading training data
df = pd.read_csv(os.path.join(DATA_PATH, TRAIN_FILENAME))
df.head()
```

In case you may ask about [metadata](https://www.kaggle.com/c/titanic/data), here is some useful points to consider:

- **_PassengerId_:** reference of passenger id registered on the trip;
- **_Survived_:** target variable indicating passenger's survival (1=yes, 0=no);
- **_Pclass_:** passenger's social class (ticket's reference) (1=high, 2=medium ou 3=low);
- **_Name_:** passenger's name;
- **_Sex_:** passenger's gender;
- **_Age_:** passenger's age;
- **_SibSp_:** total of passenger's siblings / spouses aboard the ship;
- **_Parch_:** total of passenger's parents / children aboard the ship;
- **_Ticket_:** passenger's ticket reference number;
- **_Fare_:** amount paid of passenger on the ticket bought;
- **_Cabin_:** passenger's cabin reference number;
- **_Embarked_:** port of embarkation(C=Cherbourg, Q=Queenstown, S=Southampton).

# Data Prep: Pipeline Construction

After going through a lot of explorations using visual concepts, we got useful insights for making the modeling work easier and more intuitive. By this time, the topics below will present a data prep pipeline construction using tools of **pycomp** package like the `ml.transformers` module and other functions and classes by **sklearn**.

Let's beging with a complete overview from the data given by `data_overview()` function of `pycomp.viz.insights` module.

```python
# Data overview
data_overview(df=df)
```

The table above shows rich information for each feature on the dataset abd by this, we can see that `Cabin`, `Age` and `Embarked` have null data that we must handle during the data prep process. By the way, this table can be a really good information for making a earlier step: feature selection.

## Feature Selection

```python
# Initial features
TO_DROP = ['PassengerId', 'Name', 'Ticket', 'Cabin']
INITIAL_FEATURES = list(df.drop(TO_DROP, axis=1).columns)
```

So, let's start our pipeline by importing the `pycomp.ml.transformers` module and using some classes already built for making this process easiers. The first tranformer we will use helps selecting a set of pre-defined features.

```python
# Importing class
from pycomp.ml.transformers import FiltraColunas

# Applying transformer
selector = FiltraColunas(features=INITIAL_FEATURES)
df_filtered = selector.fit_transform(df)

# Results
print(f'Shape of original dataset: {df.shape}')
print(f'Shape of dataset after selection: {df_filtered.shape}')
df_filtered.head()
```

The application of `FiltraColunas` class resulted on a 4 columns reduction on our original dataframe.

## Dropping Duplicates

The handling of duplicate data is an important step to be taken during the base preparation phase. Among the existing possibilities, the elimination of duplicates is one of the most preferable, since it brings a direct way to reduce data redundancy, thus helping the model to convert faster to the minimum computed error.

```python
# Importing class
from pycomp.ml.transformers import EliminaDuplicatas

# Applying transformer
dup_dropper = EliminaDuplicatas()
df_nodup = dup_dropper.fit_transform(df_filtered)

# Results
print(f'Total of duplicates before: {df_filtered.duplicated().sum()}')
print(f'Total of duplicates after: {df_nodup.duplicated().sum()}')
```

___
With sections 3.1 and 3.2 we can build a `initial_pipeline` with common transformers, i.e., transformers that can be applied on the entire dataset
___

```python
# Building a initial pipeline
from sklearn.pipeline import Pipeline

initial_pipeline = Pipeline([
    ('selector', FiltraColunas(features=INITIAL_FEATURES)),
    ('dup_dropper', EliminaDuplicatas())
])

# Applying this pipeline into the original data
df_prep = initial_pipeline.fit_transform(df)
print(f'Shape of original dataset: {df.shape}')
print(f'Shape of prep dataset: {df_prep.shape}')
```

## Specific Pipelines

Amazing! In a single object called `initial_pipeline` it is possible to encapsulate different _transformers_ in a single data preparation block. The `fit_transform ()` method of this object receives a raw input base and applies, sequentially, the other transforming methods of each of the classes present in the created pipeline.

The result obtained is a database initially prepared for more advanced analyzes. For this, a separation of the database by primitive type will be proposed below, thus creating two strands of preparation: one for numerical data and another for categorical data.

```python
# Separando dados em X e y
X = df_prep.drop('Survived', axis=1)
y = df_prep['Survived'].values

# Separando colunas numéricas e categóricas
num_features = [col for col, dtype in X.dtypes.items() if dtype != 'object']
cat_features = [col for col, dtype in X.dtypes.items() if dtype == 'object']

# Separando bases
X_num = X[num_features]
X_cat = X[cat_features]

# Verificando
print(f'Numerical features:\n{num_features}')
print(f'\nCategorical features:\n{cat_features}')
```

### Numerical Pipeline

At this time of preparation, it is necessary to analyze numerical and categorical data separately in order to understand the best way to get around some specific situations for each primitive type. In the case of numerical data, it is necessary to investigate the presence and origin of possible null data present in these columns.

```python
# Null data on numerical features
X_num.isnull().sum()
```

```python
# Creating a numerical pipeline that just fills null data
from sklearn.impute import SimpleImputer

# Building a numerical pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])

# Executing the pipeline
X_num_prep = num_pipeline.fit_transform(X_num)

# Results
print(f'Null data before imputer: {X_num.isnull().sum().sum()}')
print(f'Null data after imputer: {pd.DataFrame(X_num_prep).isnull().sum().sum()}')
```

### Categorical Pipeline

For the categorical step of the pipeline, we can investigate:

1. Presence of null data on categorical features;
2. Application of encoding for handling categorical entries

```python
# Null data on categorical features
X_cat.isnull().sum()
```

Well, we have 2 null data for `Embarked` variable. It's too little but we can handle it using the same `SimpleImputer` class we used on numerical pipeline but with a different `strategy`. For encoding, we can use the `DummiesEncoding()` class given by `pycomp` package.

```python
# Importing modules
from pycomp.ml.transformers import DummiesEncoding

# Building a categorical pipeline
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoding', DummiesEncoding(cat_features_ori=cat_features))
])

# Applying pipeline
X_cat_prep = cat_pipeline.fit_transform(X_cat)

# Results
print(f'Null data on X_cat_prep: \n{X_cat_prep.isnull().sum()}')
X_cat_prep.head()
```

After the construction and execution of the categorical pipeline, we can see the transformations applied on our data, mainly by the fact that we could use imputer to fill null data and use `pycomp` to encode the data.

With the steps 3.3.1 and 3.3.2 we can now build a full pipeline for encapsulate `numerical` and `categorical` transformations in only one step.

## Complete Pipeline

The code below uses sklearns's `ColumnTransformer` class for building a complete pipeline joining numerical and categorical pipelines already built. The result is an object that can receive an input `X` data and returns a prepared array for a machine learning model using the transformers inside each pipeline.

```python
# Importing libraries
from sklearn.compose import ColumnTransformer

# Building a complete pipeline
prep_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])

# Applying pipeline
X_prep = prep_pipeline.fit_transform(X)

# Results
print(f'Dimensões de X: {X.shape}')
print(f'Dimensões de X_prep: {X_prep.shape}')
```

One important step is to return a set of the final features to be used on a machine learning model. As long as we built a categorical pipeline with `dummies encoding` step, it's expected that the final set of features changes. So, the `DummiesEncoding()` class from `pycomp` handles it by saving an attribute after the encoding step with the final features after the process. Let's return it and create a final set of model features.

```python
# Returning features after encoding
encoder = prep_pipeline.named_transformers_['cat'].named_steps['encoding']
cat_encoded_features = encoder.features_after_encoding

# Creating a definite set o features
model_features = num_features + cat_encoded_features
print(f'Sanity check - quantidade de features (deve ser igual a {X_prep.shape[1]}): {len(model_features)}')
```

# Modeling: Predicting Survival

Once again, the package `pycomp` will help a lot on the task of building a flow for training and evaluating some candidate machine learning models for predicting the survival of Titanic's passengers.

For this to be done, we should import the functions and classes of `pycomp.ml.trainer` module and handle the preparation of the estimators and the hyperparameters for each estimator to be used on random search.

```python
# Importing classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Generating objects
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
forest = RandomForestClassifier()

# Defining hyperparameters search space
# Logistic Regression
logreg_param_grid = {
    'C': np.linspace(0.1, 10, 20),
    'penalty': ['l1', 'l2'],
    'class_weight': ['balanced', None],
    'random_state': [42],
    'solver': ['liblinear']
}

# Decision Trees
dtree_param_grid = {
    'criterion': ['entropy', 'gini'],
    'max_depth': [3, 5, 10, 20],
    'max_features': np.arange(1, X_prep.shape[1]),
    'class_weight': ['balanced', None],
    'random_state': [42]
}

# Random Forest
forest_param_grid = {
    'bootstrap': [True, False],
    'max_depth': [3, 5, 10, 20, 50],
    'n_estimators': [50, 100, 200, 500],
    'random_state': [42],
    'max_features': ['auto', 'sqrt'],
    'class_weight': ['balanced', None]
}
```

```python
# Preparando set de classificadores
set_classifiers = {
    'LogisticRegression': {
        'model': logreg,
        'params': logreg_param_grid
    },
    'DecisionTrees': {
        'model': dtree,
        'params': dtree_param_grid
    },
    'RandomForest': {
        'model': forest,
        'params': forest_param_grid
    }
}
```

Once we prepared the classifiers and its hyperparemeters, it's possible to use the class `ClassificadorBinario` from `pycomp.ml.trainer` module for training and evaluating all the models in just one step:

```python
# Importing class
from pycomp.ml.trainer import ClassificadorBinario

# Generating object
trainer = ClassificadorBinario()

# Training models
trainer.training_flow(set_classifiers, X_prep, y, X_test=None, y_test=None, features=model_features, random_search=True)
```
