# Prep: Construção de Pipelines

Após uma longa jornada na sessão de análise exploratória, reunimos valiosos insights que podem tornar o trabalho de modelagem muito mais fácil e intuitivo. Assim, os próximos tópicos desse notebook irão tratar de um passo extremamente importante para o treinamento de um modelo de machine learning: a preparação da base de dados.

Para facilitar este trabalho, será utilizada uma ferramenta muito poderosa construída dentro do pacote `pycomp`. Trata-se do módulo `pycomp.ml.transformers`, sendo este composto por classes prontas capazes de realizar uma série de atividades e transformações em uma base de dados, além de uma completa integração com `Pipelines` de data prep.

Antes de iniciar os passos referentes a essas transformações, vamos aplicar uma outra função igualmente poderosa do módulo `pycomp.ml.insights` capaz de retornar um overview geral de uma base de dados.

```python
# Data overview
data_overview(df=df)
```

A função `data_overview()` retorna um DataFrame contendo informações relevantes sobre cada uma das colunas presentes em uma base de dados passada como input. O resultado acima mostra parâmetros como a quantidade de dados nulos, o tipo primitivo e a quantidade de entradas categóricas (em caso de atributos categóricos) para cada uma das colunas da base.

Como insight, é possível analisar que, considerando um viés preditivo, nem todas as colunas da base serão utilizadas para a predição de sobrevivência dos passageiros do navio. Na prática, colunas como `Cabin`, `PassengerId`, `Ticket` e `Name`, por exemplo, não possuem significado prático para uma modelagem preditiva, sendo responsáveis apenas por informações chave de passageiros ou indicadores únicos de registros.

Dessa forma, o primeiro passo que poderia ser dado nessa sessão de Data Prep é aplicar um processo de _feature selection_ na base de dados, filtrando apenas os atributos candidatos a modelagem preditiva.

## Pipeline Inicial

### Feature Selection

```python
# Definindo set inicial de features
TARGET = 'Survived'
TO_DROP = ['PassengerId', 'Name', 'Ticket', 'Cabin']
INITIAL_FEATURES = list(df.drop(TO_DROP, axis=1).columns)
```

Definidas as variáveis presentes no set inicial de features (variável `INITIAL_FEATURES`), é possível utilizar a classe `FiltraColunas()` do módulo `pycomp.ml.transformers` para aplicar o processo de _feature selection_ em atributos pré-selecionados.

```python
# Importando classe
from pycomp.ml.transformers import FiltraColunas

# Aplicando transformador
selector = FiltraColunas(features=INITIAL_FEATURES)
df_slct = selector.fit_transform(df)

# Resultados
print(f'Dimensões do dataset original: {df.shape}')
print(f'Dimensões do dataset após seleção de features: {df_slct.shape}')
df_slct.head()
```

Pelas dimensões resultantes, percebe-se que a aplicação da classe `FiltraColunas` resultou na redução de 4 colunas da base original.

### Dados Duplicados

O tratamento de dados duplicados é um importante passo na preparação da base para o treinamento de modelos. Isso pois, eliminar duplicatas significa, também, eliminar redundância na base, permitindo aos modelos de machine learning uma convergência mais rápida aos mínimos locais/globais.

Para realizar essa tarefa, será utilizada a classe `EliminaDuplicatas` do módulo `pycomp.ml.transformers` que, por sua vez, é responsável por simplesmente eliminar os registros duplicados de uma base de dados passada como input.

```python
# Importing class
from pycomp.ml.transformers import EliminaDuplicatas

# Applying transformer
dup_dropper = EliminaDuplicatas()
df_nodup = dup_dropper.fit_transform(df_slct)

# Results
print(f'Total of duplicates before: {df_slct.duplicated().sum()}')
print(f'Total of duplicates after: {df_nodup.duplicated().sum()}')
```

### Dados de Treino e de Teste

Finalizando o que poderíamos chamar de _pipeline inicial_ do projeto, temos um passo importante responsável por separar a base de dados em treino e teste. Pensando em um futuro passo de modelagem, avaliar o resultado em diferentes bases é extremamente importante para tomar decisões a respeito da melhor solução prática a ser colocada em produção.

Para isso, utilizaremos a classe `SplitDados` também do módulo `pycomp.ml.transformers` que, por sua vez, aplica essa separação na base e nos retorna dados de treino e teste devidamente separados.

```python
# Importing class
from pycomp.ml.transformers import SplitDados

# Applying transformer
splitter = SplitDados(target='Survived')
X_train, X_test, y_train, y_test = splitter.fit_transform(df_nodup)

# Results
print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
print(f'Shape of y_train: {y_train.shape}')
print(f'Shape of y_test: {y_test.shape}')
```

Com isso, encerramos a primeira etapa referente a análise de transformadores a serem aplicados em um _pipeline inicial_ da base. A ideia é aplicar essas etapas comuns a base como um todo.

A seguir, serão construídos pipelines específicos de acordo com o tipo primitivo de cada coluna. Dessa forma, teremos um _pipeline numérico_ e um _pipeline categórico_ que irão compor a segunda fase de transformação no processo de prep.

```python
# Building a initial pipeline
from sklearn.pipeline import Pipeline

initial_pipeline = Pipeline([
    ('selector', FiltraColunas(features=INITIAL_FEATURES)),
    ('dup_dropper', EliminaDuplicatas())
])

# Applying this pipeline into the original data
df_prep = initial_pipeline.fit_transform(df)
print(f'Shape of original dataset: {df.shape}')
print(f'Shape of prep dataset: {df_prep.shape}')
```

### Encoding

Um passo extremamente importante envolvendo atributos categóricos da base é a aplicação de um processo conhecido como `encoding` dos dados. A importância desse _step_ se dá pela incapacidade da maioria de modelos de machine learning em ler entradas categóricas de uma base de dados fornecida como input. Isto pois, no backgorund dos modelos mais comuns, são realizados diversos cálculos numéricos em prol da minimização de uma função custo e alcance de um erro mínimo, o que de fato fica impossibilitado na presença de variávels categóricas.

Para isso, o processo de `encoding` existe para aplicar uma espécie de "codificação" nos dados categóricos, muitas vezes os separando em diferentes colunas, uma para cada entrada diferente. Neste contexto, podemos utilizar a classe `DummiesEncoding()` do módulo `pycomp.ml.transformers` que, por sua vez, é responsável por aplicar esse processo de forma automática.

```python
# Importando classe
from pycomp.ml.transformers import DummiesEncoding

# Criando objeto e aplicando transformador
encoder = DummiesEncoding(dummy_na=False)
X_train_cat_enc = encoder.fit_transform(X_train_cat_imp)

# Resultados
print(f'Shape before encoding: {X_train_cat_imp.shape}')
print(f'Shape after encoding: {X_train_cat_enc.shape}')
X_train_cat_enc.head()
```

## Pipelines Completos

Após passarmos por 3 grandes blocos envolvendo a configuração e aplicação de classes específicas para a preparação da nossa base de dados, é possível consolidar esses passos em objetos nativos a partir da classe `Pipeline` do `sklearn`.

Nessa sessão, iremos construir pipelines capazes de consolidar todas as etapas detalhadas até aqui em objetos únicos de preparação. Tais objetos serão extremamente importante na reprodução desse fluxo e também na aplicação padronizada das etapas em novos dados a serem recebidos (dados de teste, por exemplo). Assim, considerando os tópicos anteriores, iremos construir:

1. **_initial_pipeline:_** pipeline responsável por receber uma base bruta de dados e aplicar transformações comuns a todos os atributos;
2. **_prep_pipeline:_** pipeline responsável por consolidar as transformações aplicadas em dados numéricos e categóricos em um único objeto.

```python
# Importando bibliotecas
from sklearn.pipeline import Pipeline

# Construindo pipeline initial
initial_pipeline = Pipeline([
    ('selector', FiltraColunas(features=INITIAL_FEATURES)),
    ('dup_dropper', EliminaDuplicatas()),
    ('splitter', SplitDados(target='Survived'))
])

# Aplicando pipeline
X_train, X_test, y_train, y_test = initial_pipeline.fit_transform(df)
print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
print(f'Shape of y_train: {y_train.shape}')
print(f'Shape of y_test: {y_test.shape}')
```

O bloco acima basicamente consolida toda a preparação detalhada em 3.1.1, 3.1.2 e 3.1.3 em um bloco único denominiado _initial_pipeline_. Vamos agora realizar um procedimento semelhante para consolidar as etapas contidas em 3.2.1, 3.3.1 e 3.3.2

```python
# Importando bibliotecas
from sklearn.compose import ColumnTransformer

# Construindo pipeline numérico
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])

# Construindo pipeline categórico
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', DummiesEncoding(dummy_na=False, cat_features_ori=cat_features))
])

# Consolidando pipelines em objeto único
prep_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])

# Aplicando pipeline de preparação
X_train_prep = prep_pipeline.fit_transform(X_train)
X_test_prep = prep_pipeline.fit_transform(X_test)
print(f'Dimensões de X_train_prep: {X_train_prep.shape}')
```

```python
# Criando set final de features do modelo
cat_encoded_features = prep_pipeline.named_transformers_['cat'].named_steps['encoder'].features_after_encoding
MODEL_FEATURES = num_features + cat_encoded_features
print(f'Total de features consideradas (deve ser igual a {X_train_prep.shape[1]}): {len(MODEL_FEATURES)}')
```

Perfeito! Neste ponto do projeto, estamos prontos para iniciar a etapa de modelagem preditiva em busca de construir um modelo capaz de retornar a probabilidade de sobrevivência dos passageiros do Titanic frente as variáveis consideradas para análise.

## Definições Iniciais

```python
# Importing classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Generating objects
logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
forest = RandomForestClassifier()

# Logistic Regression
logreg_param_grid = {
    'C': np.linspace(0.1, 10, 20),
    'penalty': ['l1', 'l2'],
    'class_weight': ['balanced', None],
    'random_state': [42],
    'solver': ['liblinear']
}

# Decision Trees
dtree_param_grid = {
    'criterion': ['entropy', 'gini'],
    'max_depth': [3, 5, 10, 20],
    'max_features': np.arange(1, len(MODEL_FEATURES)),
    'class_weight': ['balanced', None],
    'random_state': [42]
}

# Random Forest
forest_param_grid = {
    'bootstrap': [True, False],
    'max_depth': [3, 5, 10, 20, 50],
    'n_estimators': [50, 100, 200, 500],
    'random_state': [42],
    'max_features': ['auto', 'sqrt'],
    'class_weight': ['balanced', None]
}
```

```python
# Preparando set de classificadores
set_classifiers = {
    'LogisticRegression': {
        'model': logreg,
        'params': logreg_param_grid
    },
    'DecisionTrees': {
        'model': dtree,
        'params': dtree_param_grid
    },
    'RandomForest': {
        'model': forest,
        'params': forest_param_grid
    }
}
```
